```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report,recall_score,f1_score,precision_score
```

```python
# Load dataset
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# ID3 (entropy)
model = DecisionTreeClassifier(criterion="entropy")
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8))
plot_tree(model,feature_names=data.feature_names,class_names=y,filled=True,max_depth=2)
plt.title("Decision Tree using ID3")
plt.show()
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("classification_report:", classification_report(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1 Score:", f1_score(y_test, y_pred, average='macro'))
```

```python
from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load built-in dataset
data = load_diabetes()
```

```python
X = data.data
y = data.target
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# CART Regression
model = DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("MSE:", mean_squared_error(y_test,y_pred))
print("R2 Score:", r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(12,8))
plot_tree(model,feature_names=data.feature_names,filled=True,max_depth=3)

plt.title("Decision Tree using CART")
plt.show()
```
